## Split Data Train dan Test

```python
from sklearn.model_selection import train_test_split

# Memisahkan fitur dan target berdasarkan label kolom
X = df.drop('median_house_value', axis=1) # Semua kolom kecuali target
y = df['median_house_value']              # Hanya kolom target

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=20)
```

## Pipeline Preprocessing Data dan Modeling

```python
num_features = X_train.select_dtypes(include=['int64', 'float64']).columns
cat_features = ["ocean_proximity"]

print(num_features, "\n", cat_features)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import RobustScaler, OneHotEncoder
from sklearn.svm import SVR
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression, SGDRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor


numeric_pipeline = Pipeline(steps=[
        ("imputer", SimpleImputer(strategy="median")), 
        ("scaler", RobustScaler())
    ])

categorical_pipeline = Pipeline(steps=[
        ("imputer", SimpleImputer(strategy="most_frequent")),
        ("encoder", OneHotEncoder(handle_unknown="ignore"))
    ])

preprocessing_pipeline = ColumnTransformer(transformers=[
            ("num", numeric_pipeline, num_features),
            ("cat", categorical_pipeline, cat_features)    
    ])

model_pipeline = Pipeline(steps=[
    ("preprocessing", preprocessing_pipeline),
    ("model", LinearRegression())
    ])

model_pipeline2 = Pipeline(steps=[
    ("preprocessing", preprocessing_pipeline),
    ("model", SGDRegressor())
    ])

model_pipeline3 = Pipeline(steps=[
    ("preprocessing", preprocessing_pipeline),
    ("model", RandomForestRegressor(n_estimators=100, random_state=20))
])

model_pipeline4 = Pipeline(steps=[
    ("preprocessing", preprocessing_pipeline),
    ("model", GradientBoostingRegressor(n_estimators=100, random_state=20))
])
```

```python
model_pipeline.fit(X_train, y_train)

print('Train R^2 Score : %.3f'%model_pipeline.score(X_train, y_train))
print('Test R^2 Score : %.3f'%model_pipeline.score(X_test, y_test))
```

```python
model_pipeline2.fit(X_train, y_train)

print('Train R^2 Score : %.3f'%model_pipeline2.score(X_train, y_train))
print('Test R^2 Score : %.3f'%model_pipeline2.score(X_test, y_test))
```

```python
model_pipeline3.fit(X_train, y_train)

print('Train R^2 Score : %.3f'%model_pipeline3.score(X_train, y_train))
print('Test R^2 Score : %.3f'%model_pipeline3.score(X_test, y_test))
```

```python
model_pipeline4.fit(X_train, y_train)

print('Train R^2 Score : %.3f'%model_pipeline4.score(X_train, y_train))
print('Test R^2 Score : %.3f'%model_pipeline4.score(X_test, y_test))
```

Best Model adalah Random Forest dengan akurasi train melebihi 90% dan test melebih 80%

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor

param_grid = {
    'model__n_estimators': [100, 200, 300],
    'model__max_depth': [10, 20, 30, None],
    'model__min_samples_split': [2, 5, 10]
}

model_pipeline_rf = Pipeline(steps=[
    ("preprocessing", preprocessing_pipeline),
    ("model", RandomForestRegressor(random_state=20))
])

grid_search = GridSearchCV(model_pipeline_rf, param_grid, cv=5, 
                          scoring='r2', n_jobs=-1, verbose=2)
grid_search.fit(X_train, y_train)

print('Best parameters:', grid_search.best_params_)
print('Best R^2 Score:', grid_search.best_score_)
```

```python
# Use the best estimator from GridSearchCV
model_pipeline_best = grid_search.best_estimator_

# Evaluate on test set
print('Best Model Performance:')
print('Train R^2 Score : %.3f' % model_pipeline_best.score(X_train, y_train))
print('Test R^2 Score : %.3f' % model_pipeline_best.score(X_test, y_test))
```
